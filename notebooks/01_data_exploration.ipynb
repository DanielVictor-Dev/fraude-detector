```python
# ‚úÖ An√°lise Explorat√≥ria dos Dados

# Importa√ß√£o das bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Configura√ß√£o dos gr√°ficos
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# ‚úÖ 1. Carregar os Dados
df = pd.read_csv('../data/raw/dataset.csv')

# ‚úÖ 2. Verificar as primeiras linhas
print("üîç Primeiros registros do dataset:")
print(df.head())

# ‚úÖ 3. Verificar informa√ß√µes gerais do dataset
print("\n‚ÑπÔ∏è Informa√ß√µes do dataset:")
print(df.info())

# ‚úÖ 4. Verificar se existem valores nulos
print("\n‚ùì Verificando valores nulos:")
print(df.isnull().sum())

# ‚úÖ 5. Verificar estat√≠sticas descritivas
print("\nüìä Estat√≠sticas descritivas:")
print(df.describe())

# ‚úÖ 6. Verificar distribui√ß√£o da vari√°vel alvo (status_fraude)
print("\nüîç Distribui√ß√£o da vari√°vel alvo (status_fraude):")
print(df['status_fraude'].value_counts())

# ‚úÖ 7. Plotar gr√°fico de distribui√ß√£o da vari√°vel alvo
sns.countplot(data=df, x='status_fraude')
plt.title('Distribui√ß√£o da Classe - Fraude')
plt.xlabel('Status Fraude (0 = N√£o Fraude, 1 = Fraude)')
plt.ylabel('Quantidade')
plt.savefig('../outputs/figures/distribuicao_classe.png')
plt.show()

# ‚úÖ 8. Correla√ß√£o entre vari√°veis num√©ricas
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues', fmt='.2f')
plt.title('Mapa de Calor - Correla√ß√£o entre Vari√°veis Num√©ricas')
plt.savefig('../outputs/figures/correlacao_variaveis.png')
plt.show()

# ‚úÖ 9. An√°lise de algumas vari√°veis (Exemplo: Valor da transa√ß√£o)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='status_fraude', y='valor_transacao')
plt.title('Distribui√ß√£o do Valor das Transa√ß√µes por Status de Fraude')
plt.xlabel('Status Fraude (0 = N√£o Fraude, 1 = Fraude)')
plt.ylabel('Valor da Transa√ß√£o')
plt.savefig('../outputs/figures/boxplot_valor_transacao.png')
plt.show()

# ‚úÖ 10. Verificar balanceamento da vari√°vel alvo
fraude_percent = df['status_fraude'].value_counts(normalize=True) * 100
print("\nüìä Distribui√ß√£o percentual da vari√°vel alvo:")
print(fraude_percent)
```
